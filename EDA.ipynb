```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
#line plot

x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.title("Distribution")
plt.xlabel("Data1")
plt.ylabel("Data2")
plt.show()
```

```python
#Bar plot(use pandas library)

data=pd.DataFrame({
    "Class":["A","B","C","D"],
    "Student":[50,46,36,20]
})
sns.barplot(x='Class',y='Student',data=data)
plt.title("Bar plot")
plt.show()
```

class A having more students count followed by class B 

```python
#Count plot(use pandas library)
sns.countplot(x='Class',data=data)
plt.title("Count plot")
plt.show()
```

```python
#Histogram(use numpy library) 
data=np.random.randn(100)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
#KDE plot
sns.kdeplot(data,shade=True)
plt.title("KDE plot")
plt.show()
```

This graph show data is normally distributed

```python
sns.histplot(data,bins=10,kde=True)
plt.title("Histogram with KDE")
plt.show()
```
